# 1(A): find the joint probability of the people who planned to purchase and actually placed an order.
P(A)= probability of people planned to purchase product A
P(B)= probability of people actually placed order for product A
n(A)= people who planned to purchase product A= 500
n(B)= people who actually placed oreder for product A= 600
total no. of events= 2000
A intersection B= 400
so the joint probabilty pJ = n(A intersection B)/total no. of events



```python
FE= 400    #favorable events
TE= 2000   #total no. of events
PJ= round(FE / TE, 4)  # joint probabilty as mentioned in question 1 
print('1.A.joint probability of the people who planned to purchase and actually placed an order. is %1.4f' % PJ)
```

# 1.B.joint probility of the people who planned to purchase and actually placed an order, given that peole planned to purchase 
PJ2= P(B)*P(B/A), where P(B/A)= conditional probability of B, given the fact A alreadt happend


```python
FE1= 600 # favourbale ways for B event 
TE1= 2000# total no. of ways
pb= round(FE1/TE1, 4) # probabilty of people actually placed order for product A
print ('pb is %1.4lf\n'% pb)
```

```python
FE2= 200 # 
TE2= 1600
pc= round(FE2/TE2,4)# conditional probabilty of B, givan the fact A already happened
print('pc is %1.4f'%pc)
```

```python
pd = round(pb*pc,4) # joint probility of the people who planned to purchase and actually placed an order, given that peole planned to purchase
print('1.B.joint probility of the people who planned to purchase and actually placed an order, given that peole\n\
planned to purchase. is %1.4f'%pd)
```

```python
n=10 # random samples of 10 items
p= 0.05 # percentage failure rate for manufacture items
k= np.arange(0,11)
binomial= stats.binom.pmf(k,n,p)
print (binomial)
```
